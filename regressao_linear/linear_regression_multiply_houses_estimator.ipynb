```python
import pandas as pd
base = pd.read_csv('house_prices.csv')
base.head()
```

```python
base.columns
```

```python
base.drop(['id', 'date', 'sqft_living15', 'sqft_lot15'], axis=1, inplace=True)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler_x = MinMaxScaler()
base[['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 
'floors', 'waterfront', 'view', 'condition', 'grade', 'sqft_above', 'sqft_basement', 'yr_built', 
'yr_renovated', 'zipcode', 'lat', 'long']] = scaler_x.fit_transform(base[['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 
'floors', 'waterfront', 'view', 'condition', 'grade', 'sqft_above', 'sqft_basement', 'yr_built', 
'yr_renovated', 'zipcode', 'lat', 'long']])
```

```python
base.head()
```

```python
scaler_y = MinMaxScaler()
base[['price']] = scaler_y.fit_transform(base[['price']])
```

```python
base.head()
```

```python
x = base.drop('price', axis = 1)
x.head()
```

```python
y = base.price
y.head()
type(y)
```

```python
columns_predictors = base.columns[3:]
columns_predictors
```

```python
import tensorflow as tf
```

```python
columns  = [tf.feature_column.numeric_column(key = name) for name in columns_predictors]
```

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size = 0.3)
```

```python
def make_input_fn(x, y = None, num_epochs = None, shuffle = True, batch_size = 32):
    def input_function():
        ds = tf.data.Dataset.from_tensor_slices((dict(x), y))
        if shuffle:
            ds = ds.shuffle(1000)
        ds = ds.batch(batch_size).repeat(num_epochs)
        return ds
    return input_function
```

```python
ds = tf.data.Dataset.from_tensor_slices((dict(train_x), train_y))
ds
```

```python
train_function = make_input_fn(train_x, train_y)
```

```python
train_function()
```

```python
regressor = tf.estimator.LinearRegressor(feature_columns = columns)
```

```python
regressor.train(input_fn = train_function, steps = 10000)
```

```python
train_metrics = regressor.evaluate(input_fn = train_function, steps = 10000)
```

```python
test_function = make_input_fn(test_x, test_y, num_epochs = 10000, shuffle = False)
```

```python
test_metrics = regressor.evaluate(input_fn = test_function, steps = 10000)
```

```python
print(test_metrics)
```

```python
predict_function = make_input_fn(test_x, num_epochs = 1, shuffle = False)
```

```python
predicts = regressor.predict(input_fn = predict_function)
```

```python
predicts_results = []
for predict in regressor.predict(input_fn = predict_function):
    predicts_results.append(predict['predictions'])
```

```python
import numpy as np
predicts_results = np.asarray(predicts_results).reshape(-1, 1)
```

```python
predicts_results = scaler_y.inverse_transform(predicts_results)
y_results = test_y.values.reshape(-1, 1)
y_results = scaler_y.inverse_transform(y_results)
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_results, predicts_results)
```
